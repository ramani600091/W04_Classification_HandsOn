```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def getRows(ds):
    data=pd.read_csv(ds)
    data
    dataset=pd.get_dummies(data,dtype=int,drop_first=True)
    independent=dataset[[  'Age','EstimatedSalary','Gender_Male']]
    dependent=dataset[["Purchased"]]
    dataset=dataset.drop("User ID",axis=1)
    shape=independent.shape
    return independent,dependent,shape
```

```python
def getSVMLinearClassificaiton(ind,dep,vkernel):
    independent = ind
    dependent   = dep
    ##-SPLIT TRAIN & TEST
    from sklearn.model_selection import train_test_split as tts
    X_Train,X_Test,Y_Train,Y_Test=tts(independent,dependent,test_size=1/3,random_state=0)
    
    ##STANDARDIZATION
    from sklearn.svm import SVC
    classifier = SVC(kernel = vkernel, random_state = 0)
    classifier.fit(X_Train, Y_Train)

    y_pred = classifier.predict(X_Test)

    ##CLASSIFICATION
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(Y_Test, y_pred)     
    
    from sklearn.metrics import classification_report
    clf_report = classification_report(Y_Test, y_pred)  
    
    return clf_report,cm,y_pred
```

```python
def getDecisionTreeClassification(ind,dep,vkernel):
    independent = ind
    dependent   = dep
    ##-SPLIT TRAIN & TEST
    from sklearn.model_selection import train_test_split as tts
    X_Train,X_Test,Y_Train,Y_Test=tts(independent,dependent,test_size=1/3,random_state=0)
    
    ##STANDARDIZATION
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion = vkernel, random_state = 0)

    classifier.fit(X_Train, Y_Train)

    y_pred = classifier.predict(X_Test)

    ##CLASSIFICATION
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(Y_Test, y_pred)     
    
    from sklearn.metrics import classification_report
    clf_report = classification_report(Y_Test, y_pred)  
    
    return clf_report,cm,y_pred    
```

```python
def getKNearestNeighbor(ind,dep,vkernel):
    independent = ind
    dependent   = dep
    ##-SPLIT TRAIN & TEST
    from sklearn.model_selection import train_test_split as tts
    X_Train,X_Test,Y_Train,Y_Test=tts(independent,dependent,test_size=1/3,random_state=0)
    
    ##STANDARDIZATION
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors = 7, metric = 'minkowski', p = 2)
    classifier.fit(X_Train, Y_Train)           
    y_pred = classifier.predict(X_Test)

    ##CLASSIFICATION
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(Y_Test, y_pred)     
    
    from sklearn.metrics import classification_report
    clf_report = classification_report(Y_Test, y_pred)  

    return clf_report,cm,y_pred        
```
